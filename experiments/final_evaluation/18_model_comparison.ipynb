```python
import pandas as pd

tracking = pd.read_csv("results/model_tracking.csv")

tracking = tracking.drop_duplicates(
    subset=["model_id"],
    keep="last"
)

tracking["accuracy_gap"] = (
    tracking["train_accuracy"] - tracking["test_accuracy"]
)

leaderboard = tracking.sort_values(
    by="f1_score",
    ascending=False
)

leaderboard[
    [
        "model_id",
        "model_name",
        "features",
        "train_accuracy",
        "test_accuracy",
        "accuracy_gap",
        "precision",
        "recall",
        "f1_score",
        "model_path"
    ]
].head(10)
```
