This notebook follows the same structure as the existing Bag of Words experiments:

```python
import os
import sys

os.chdir("/Users/karima/Ironhack-challenges/fake-news-nlp-classification")
sys.path.append(os.getcwd())

print(os.getcwd())
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

from src.data_loader import load_data, prepare_features_and_labels, split_data
from src.preprocessing import preprocess_data
from src.feature_extraction import create_tfidf_bigram_features
from src.evaluator import evaluate_model, save_confusion_matrix, save_metrics_plot
from src.experiment_tracker import save_experiment_result
from src.model_manager import save_model, save_vectorizer
from src.config import MODELS_DIR
```

```python
# Class constants
MODEL_ID = "exp_09"
MODEL_NAME = "Logistic Regression + TF-IDF + Bigrams"
FEATURES = "TF-IDF + n-grams (unigrams and bigrams)"
PREPROCESSING = "Lowercase + HTML removal + stopword removal + lemmatization"
ALGORITHM = "Logistic Regression"
NOTES = "Baseline Logistic Regression model using TF-IDF unigram and bigram features."
MODEL_PATH = MODELS_DIR / "tfidf_bigrams_logistic_regression.pkl"
```

```python
# Load the dataset
data = load_data()
```

```python
# Extract features X and labels y
X, y = prepare_features_and_labels(data)
```

```python
# Split the data into 80% train and 20% test
X_train, X_test, y_train, y_test = split_data(X, y)
```

```python
# Data cleaning, tokenization and lemmatization
X_train_clean = preprocess_data(X_train)
X_test_clean = preprocess_data(X_test)
```

```python
# Create TF-IDF + n-grams (unigrams and bigrams) features
X_train_features, X_test_features, vectorizer = create_tfidf_bigram_features(
    X_train_clean,
    X_test_clean
)
```

```python
# Train the model
model = LogisticRegression(
    random_state=42,
    max_iter=1000
)

model.fit(X_train_features, y_train)
```

```python
# Evaluate the model
metrics, predictions = evaluate_model(
    model,
    X_test_features,
    y_test
)
```

```python
# Save the model and vectorizer
model_path = save_model(
    model,
    "tfidf_bigrams_logistic_regression.pkl"
)

vectorizer_path = save_vectorizer(
    vectorizer,
    "tfidf_bigrams_logistic_regression_vectorizer.pkl"
)
```

```python
# Save and show metrics plot
metrics_plot_path = save_metrics_plot(
    metrics,
    "09_baseline_tfidf_bigrams_logistic_regression_metrics"
)
```

```python
# Save and show confusion matrix
confusion_matrix_path = save_confusion_matrix(
    y_test,
    predictions,
    "09_baseline_tfidf_bigrams_logistic_regression_confusion_matrix"
)
```

```python
# Train accuracy
train_prediction = model.predict(X_train_features)

train_accuracy = accuracy_score(
    y_train,
    train_prediction
)
```

```python
# Save experiment result
experiment = {
    "model_id": MODEL_ID,
    "model_name": MODEL_NAME,
    "features": FEATURES,
    "preprocessing": PREPROCESSING,
    "algorithm": ALGORITHM,
    "train_accuracy": train_accuracy,
    "test_accuracy": metrics["test_accuracy"],
    "precision": metrics["precision"],
    "recall": metrics["recall"],
    "f1_score": metrics["f1_score"],
    "notes": NOTES,
    "model_path": MODEL_PATH,
}

tracking = save_experiment_result(**experiment)

tracking.tail()
```
